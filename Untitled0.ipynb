```python
# je definis mon RMSE car ce n'est pas integre par defaut dans Keras
import tensorflow.keras.backend as K
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))
```

```python
# Compilation

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',           # en rgression on garde souvent MSE comme loss
    metrics=[rmse]        # on suit le RMSE en plus
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

#2. créer un callbacks EarlyStopping basé sur RMSE
early_stopping = EarlyStopping(
    monitor='val_loss',               #surveiller MSE fonction de perte sur le jeu de validation
    patience=5,                       #patience = 5 époques sans amelioration
    restore_best_weights=True,        #restaurer les meilleurs poids
    mode='min',                       #"min" signfie que nous voulons que RMSE diminue
    verbose=1
)

lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    verbose=1,
    mode='min',
    min_lr=1e-6
)

#sauvegarde le meilleur modéle
checkpoint = ModelCheckpoint(
    filepath='/content/drive/MyDrive/DL/DATASETS/Audi/best_model.keras',
    monitor='val_loss',
    save_best_only=True,     # garde seulement le meilleur
    mode='min',
    verbose=1
)
```

```python
# Modele keras utilise des flaot 32(conversion)
X_train = X_train.astype('float32')
X_val = X_val.astype('float32')
X_test = X_test.astype('float32')

Y_train = Y_train.astype('float32')
Y_val = Y_val.astype('float32')
Y_test = Y_test.astype('float32')
```

```python
#7. Entrainement
# On entraine 80% du X_train et on teste sur X_val, Y_val(attention ne pas toucher au X_test --> data_leaka)

history = model.fit(
    X_train,Y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, Y_val),
    callbacks=[early_stopping, lr_scheduler, checkpoint],
    verbose=1
)
```

```python
# calculer R2 score et val_loss sur rmse
from sklearn.metrics import r2_score
Y_pred = model.predict(X_test)
r2 = r2_score(Y_test, Y_pred)
print(f"R2 score: {r2:.4f}")

val_loss = np.sqrt(np.mean(np.square(Y_test - Y_pred.flatten())))
print(f"Validation loss(RMSE): {val_loss:.4f}")
```
